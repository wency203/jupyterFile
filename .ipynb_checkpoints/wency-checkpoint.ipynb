```python
%matplotlib inline
import matplotlib.pyplot as plt
x=np.linspace(-20,20,10)
y=x**3+2*x**2+6*x+5
plt.plot(x,y,marker="o")
```

```python
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
data=make_blobs(n_samples=200,centers=2,random_state=8)
x,y=data
plt.scatter(x[:,0],x[:,1],c=y,cmap=plt.cm.spring,edgecolor='r')
#plt.show()

import numpy as np
clf=KNeighborsClassifier()
clf.fit(x,y)
x_min,x_max=x[:,0].min()-1,x[:,0].max()+1
y_min,y_max=x[:,0].min()-1,x[:,0].max()+1
xx,yy=np.meshgrid(np.arange(x_min,x_max,.02),
                 np.arange(y_min,y_max,.02))
Z=clf.predict(np.c_[xx.ravel(),yy.ravel()])
Z=Z.reshape(xx.shape)
plt.pcolormesh(xx,yy,Z,cmap=plt.cm.Pastel1)
plt.scatter(x[:,0],x[:,1],c=y,cmap=plt.cm.spring,edgecolor='r')
plt.xlim(xx.min(),xx.max())
plt.ylim(yy.min(),yy.max())
plt.title("Classifier:KNN")

plt.scatter(6.75,4.82,marker='*',c='red',s=200)


plt.show()
```

```python
data2=make_blobs(n_samples=500,centers=5,random_state=8)
x2,y2=data2
plt.scatter(x2[:,0],x2[:,1],c=y2,cmap=plt.cm.spring,edgecolor='r')

clf=KNeighborsClassifier()
clf.fit(x2,y2)
x_min,x_max=x2[:,0].min()-1,x2[:,0].max()+1
y_min,y_max=x2[:,0].min()-1,x2[:,0].max()+1
xx,yy=np.meshgrid(np.arange(x_min,x_max,.02),
                 np.arange(y_min,y_max,.02))
Z=clf.predict(np.c_[xx.ravel(),yy.ravel()])
Z=Z.reshape(xx.shape)
plt.pcolormesh(xx,yy,Z,cmap=plt.cm.Pastel1)
plt.scatter(x2[:,0],x2[:,1],c=y2,cmap=plt.cm.spring,edgecolor='r')
plt.xlim(xx.min(),xx.max())
plt.ylim(yy.min(),yy.max())
plt.title("Classifier:KNN")

plt.show()

print('\n\n\n',"代码运行结果：\n",'=================\n')
print('模型正确率：{:.2f}'.format(clf.score(x2,y2)))
print('\n==============')
```

```python
from sklearn.datasets import make_regression
x,y=make_regression(n_features=1,n_informative=1,noise=50,random_state=8)
plt.scatter(x,y,c='orange',edgecolor='r')

from sklearn.neighbors import KNeighborsRegressor
reg=KNeighborsRegressor(n_neighbors=10)#n_neighbors参数为空时，默认为5
reg.fit(x,y)
z=np.linspace(-3,3,200).reshape(-1,1)
plt.scatter(x,y,c='orange',edgecolor='r')
plt.plot(z,reg.predict(z),c='c',linewidth=2)
plt.title('KNN Regressor')


plt.show()

print('模型评分：{:.2f}'.format(reg.score(x,y)))
```

```python
from sklearn.datasets import load_wine
wine_dataset=load_wine()

#print("红酒数据集中的键：\n{}".format(wine_dataset.keys()))
#print("\n数据概况：{}".format(wine_dataset['data'].shape))
#print("\n数据集描述：\n{}".format(wine_dataset['DESCR']))

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(
wine_dataset['data'],wine_dataset['target'],random_state=0)
# print('\nX_train shape:{}'.format(X_train.shape))
# print('\nX_test shape:{}'.format(X_test.shape))
# print('\ny_train shape:{}'.format(y_train.shape))
# print('\ny_test shape:{}'.format(y_test.shape))

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)

print('=============')
knn.fit(X_train,y_train)
# print(knn)
print('测试数据集得分：{:.2f}'.format(knn.score(X_test,y_test)))

X_new=np.array([[13.2,2.77,2.51,18.5,96.6,1.04,2.55,0.57,
                 1.47,6.2,1.05,3.33,820]])#两层中括号
prediction=knn.predict(X_new)
print("预测新红酒的分类为：{}".format(wine_dataset['target_names'][prediction]))
```

```python
x=np.linspace(-15,15,2)#linspace(x1,x2,n)返回包含 x1 和 x2 之间的 n 个等间距点的行向量
y=0.5*x+3
plt.plot(x,y,c="orange")
plt.title('Straight Line')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
X=[[1],[4]]
y=[3,5]
lr=LinearRegression().fit(X,y)
z=np.linspace(0,5,20)
plt.scatter(X,y,s=800)
plt.plot(z,lr.predict(z.reshape(-1,1)),c='k')
plt.title('Straight Line')
plt.show()

print('\n\n\n直线方程为：\n================\n')
print('y={:.3f}'.format(lr.coef_[0]),'x','+{:.3f}'.format(lr.intercept_))
print('\n================')
```
